```python
#loading Mv values
Mv = np.loadtxt('Table_Mass_0.6.txt',usecols=(3),unpack=True)
print(Mv)

gds = Table.read('/Users/tanshi/Desktop/SIP/gds.hdf5')
```

```python
# mv = -5logD + 5 + Mv
mv_100 = -5*np.log10(100) + 5 + Mv
mv_200 = -5*np.log10(200) + 5 + Mv
mv_500 = -5*np.log10(500) + 5 + Mv
mv_1000 = -5*np.log10(1000) + 5 + Mv
mv_3000 = -5*np.log10(3000) + 5 + Mv
mv_5000 = -5*np.log10(5000) + 5 + Mv
mv_10 = -5*np.log10(10) + 5 + Mv
```

```python
GR = np.loadtxt('Table_Mass_0.6.txt', usecols=(16),unpack=True)
RI = np.loadtxt('Table_Mass_0.6.txt', usecols=(8),unpack=True)
GI = GR + RI

print(gds['ID'][152])
flux_606 = gds['ACS_F606W_FLUX'][152]
mag_606 = -2.5*np.log10(flux_606)+23.9

flux_814 = gds['ACS_F814W_FLUX'][152]
mag_814 = -2.5*np.log10(flux_814)+23.9

mag606_mag814 = mag_606 - mag_814

fig=plt.figure(figsize=(10,10))
plt.scatter(GI,mv_100,c='blue',edgecolors='none',s=20,alpha=0.5)
plt.scatter(GI,mv_200,c='gray',edgecolors='none',s=20,alpha=0.5)
plt.scatter(GI,mv_500,c='red',edgecolors='none',s=20,alpha=0.5)
plt.scatter(GI,mv_1000,c='green',edgecolors='none',s=20,alpha=0.5)
plt.scatter(GI,mv_3000,c='orange',edgecolors='none',s=20,alpha=0.5)
plt.scatter(GI,mv_5000,c='yellow',edgecolors='none',s=20,alpha=0.5)
plt.scatter(GI,mv_10,c='pink',edgecolors='none',s=20,alpha=0.5)


plt.scatter(mag606_mag814,20.93, c='black',edgecolors='none',s=30,alpha=0.5)


plt.ylim(25,-8)
plt.xlabel('G-I')
plt.ylabel('mv(pc)')
#plt.savefig(out_dir+'plotting_mv_whitedwarf.pdf')
```
