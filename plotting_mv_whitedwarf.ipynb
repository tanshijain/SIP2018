```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
out_dir = '/Users/tanshi/Desktop/SIP/'
```

```python
#loading Mv values
Mv = np.loadtxt('Table_Mass_0.6.txt',usecols=(3),unpack=True)
print(Mv)
```

```python
# mv = -5logD + 5 + Mv
mv_100 = -5*np.log10(100) + 5 + Mv
mv_200 = -5*np.log10(200) + 5 + Mv
mv_500 = -5*np.log10(500) + 5 + Mv
mv_1000 = -5*np.log10(1000) + 5 + Mv


print(mv_100)
```

```python
UB = np.loadtxt('Table_Mass_0.6.txt', usecols=(4),unpack=True)
print(UB)

BV = np.loadtxt('Table_Mass_0.6.txt', usecols=(5),unpack=True)
print(BV)

UV = UB + BV
print(UV)
```

```python
fig=plt.figure(figsize=(10,10))
plt.scatter(UV,mv_100,c='blue',edgecolors='none',s=20,alpha=0.5)
plt.scatter(UV,mv_200,c='gray',edgecolors='none',s=20,alpha=0.5)
plt.scatter(UV,mv_500,c='red',edgecolors='none',s=20,alpha=0.5)
plt.scatter(UV,mv_1000,c='green',edgecolors='none',s=20,alpha=0.5)

plt.xlim(-2,2)
plt.ylim(13,0)
plt.xlabel('U-V')
plt.ylabel('mv(pc)')
plt.savefig(out_dir+'plotting_mv_whitedwarf.pdf')
```
