```python
#loading Mv values
Mv = np.loadtxt('Table_Mass_0.7.txt',usecols=(3),unpack=True)
print(Mv)
```

```python
# mv = -5logD + 5 + Mv
mv_100 = 5*np.log10(100) - 5 + Mv
mv_200 = 5*np.log10(200) - 5 + Mv
mv_500 = 5*np.log10(500) - 5 + Mv
mv_1000 = 5*np.log10(1000) - 5 + Mv

print(mv_100)
```

```python
UB = np.loadtxt('Table_Mass_0.7.txt', usecols=(4),unpack=True)
print(UB)

BV = np.loadtxt('Table_Mass_0.7.txt', usecols=(5),unpack=True)
print(BV)

UV = UB + BV
print(UV)
```
